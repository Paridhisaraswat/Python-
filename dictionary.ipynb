```python
# Merge two dictionaries and sum the values of common keys

d1 = {'a': 100, 'c': 300 ,'b': 200}
d2 = {'a': 300, 'b': 200, 'd': 400}
d = {}
for i in d1:
  for j in d2 :
    if(i == j):
      d[i] = d1[i] +d2[j]
    elif((i not in d) and (i not in d2)):
      d[i] = d1[i]
    elif((j not in d)and (j not in d1)):
      d[j] = d2[j]
print(d)
```

```python
# Write a program to remove keys with duplicate values from a dictionary
d = {'a': 10,'b': 20,'c': 10,'d': 30,'e': 20,'f': 40}
count ={}
for i in d.values():
  if i not in count :
    count[i] = 1
  elif i in count :
    count[i] = count[i] + 1
print(count)
# to store unique items only
d1 = {}
for key , val in d.items():
  if count[val] == 1:
    d1[key] = val
print(d1)
```

```python
# Sort a dictionary by its values in ascending order.
scores = {'Alice': 88,'Bob': 72,'Charlie': 95,'David': 68,'Eva': 88,'Frank': 91}
l = list(scores.values())

for i in range(len(l)):
    for j in range(i+1,len(l)):

      if (l[i] > l[j]):
        n = l[j]
        l[j] = l[i]
        l[i] = n

ascending  = {}
for i in range(len(l)):
  for key , value in scores.items():
    if(value == l[i]):
        ascending[key] = value

print(ascending)

```

```python
# Sort a dictionary by its values in descending order.
scores = {'Alice': 88,'Bob': 72,'Charlie': 95,'David': 68,'Eva': 88,'Frank': 91}
l = list(scores.values())

for i in range(len(l)):
    for j in range(i+1,len(l)):

      if (l[i] < l[j]):
        n = l[j]
        l[j] = l[i]
        l[i] = n

descending  = {}
for i in range(len(l)):
  for key , value in scores.items():
    if(value == l[i] and key not in descending):
        descending[key] = value

print(descending)

```

```python
# Write a program to find the key(s) with the maximum and minimum value in a dictionary.
scores = {'Math': 78,'Science': 92,'English': 85,'History': 69,'Geography': 92,'Computer': 88,'Art': 69}

MaxMin = {}
l = list(scores.values())
minimum = l[0]
maximum = l[0]
for i in range(len(l)):
    if (minimum > l[i]):
      minimum = l[i]
    if(maximum < l[i]):
      maximum = l[i]

for key , value in scores.items():
    if(minimum == value):
      MaxMin[key] = value
    if(maximum == value):
      MaxMin[key] = value
print(MaxMin)
```

```python
# NESTED DICTIONARY
names = ['paridhi','snehal']
subject = ['maths','science']
marks = [[93,89],[95,87]]

d ={}

for i in range(len(names)):
  inner_d ={}
  for j in range(len(subject)):
    inner_d[subject[j]] = marks[i][j]
  d[names[i]] = inner_d

print(d)
```
