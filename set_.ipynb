#set
#set is a built-in data type that stores an unordered collection of unique elements

```python
myset = set()
print(type(myset))
```

```python
myset = set({1,2,3,5,4,6,3,9,5})
print(myset)
```

```python
# to add one element
myset.add(60)
print(myset)
```

```python
myset.discard(5)
print(myset)
```

```python
# to add multiple values
myset.update([99,88,77])
print(myset)
```

```python
s = set({'apple','banana','mango','orange'})
s.add('grape')
s
```

```python
# remove element from set
s.remove("banana")
s
```

```python
# discard elements
s.discard("mango")
s
```

```python
#union - combine all the values of both the sets
set1={1,2,3}
set2 = {3,4,5}
set1.union(set2)
```

```python
#intersection - returns the values which is common in both the sets
set1= {89,32,54}
set2={32,12,67}
set1.intersection(set2)
```

```python
#difference - returns the unique values from set 1 which are not present in set 2
set1={1,2,3}
set2 = {3,4,5}
set1.difference(set2)
```

```python
#symmetric difference - returns the unique value from both the sets
set1= {89,32,54}
set2={32,12,67}
set1.symmetric_difference(set2)
```

```python
#issubset - check if both the sets hold the same data
set1={1,2,3}
set2 = {1,2,3}
set1.issubset(set2)
```

```python
# Add a list of Elements to a Set
num = ({23,56,76,879})
numbers = [32,45,84,98]
num.update(numbers)
num
```
