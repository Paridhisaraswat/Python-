```python
# to rotate a list
li = [23,43,54,2,65,1,5]
start = 0
end =len(li)-1

while(start < len(li)//2):

  a = li[start]           # temporary variable
  li[start] = li[end]     # index 0 with last value
  li[end] = a             #last index with first value
  start += 1
  end -= 1
print(li)
```

```python
# check if the starting and ending value in a list is even
# if even print true else false
#then check the 2nd and 2nd last value
l=[10,14,15,16,30,18,22]

start = 0
end = len(l)-1

while(start <= len(l)//2):
  print(start , end)
  if(l[start] % 2 == 0 and l[end] % 2 == 0):
    print("true")
  else:
    print("false")
  start += 1
  end -= 1
```

```python
#you have to get two num from the list which will create a target 21
#method 1
l= [2,3,5,8,11,16,20]
start = 0
target = 21
next = 1
while(start < len(l)):
  for i in range(next ,len(l)):
    if(l[start] + l[i] == target):
      print(l[start],l[i])
  start += 1
  next +=1
```

```python
# buy shares and sell it on incoming days of the week to make the maximum profit
# incase of no profit return 0

p=[7,1,5,3,6,4]
profit = 0
for i in range(0,len(p)):
  for j in range(i+1,len(p)):
     if(p[j] - p[i] > profit):
      profit = p[j] -p[i]

if(profit > 0 ):
  print(profit)
else:
  print("0")

```

```python
s="hey i am here hey i how are you"
duplicate=[]
unique=[]
# "i am here how are you"
newlist=s.split()
print(newlist)


for i in range(0, len(newlist) ):
  currWord=newlist[i]

  for j in range(i+1, len(newlist)):
    nextWord=newlist[j]
    if(currWord == nextWord):
      duplicate.append(currWord)


    elif(  nextWord not in unique and nextWord not in duplicate):
      unique.append(nextWord)

print("unique", unique)
```

```python
# find the sum of the elements in a list, excluding the largest
# and smallest element. Donâ€™t use max or min functions

n = [3, 2, 5, 4, 1]

sum = 0

max_num = n[0]
min_num = n[0]
for i in range(len(n)):
  if(max_num < n[i]):
    max_num = n[i]
  elif(min_num > n[i]):
    min_num = n[i]

for i in range(len(n)):
  if(n[i] != min_num and n[i] != max_num ):
    sum += n[i]
print(sum)

```

```python
# to check if the list is palindrome or not

n = [1, 2, 3, 2, 1]

start = 0
end = len(n) - 1
a  = 0
while(start < len(n)//2):
  if(n[start ] != n[end ]):
    a = 1
    break

  start += 1
  end -= 1

if(a == 1):
  print("not palindrome")
else:
  print("palindrome")
```
