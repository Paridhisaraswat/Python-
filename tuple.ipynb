```python
# Indexing
t = (10, 20, 30, 40,50)
print(t[-1])
print(t[1 + 1])       #t[1+1] ---> t[2] ----->30
print(t[4 - 1])
print(t[7%2])
print(t[2*2])
```

```python
# Scilicing
t = (1,2,3,4,5,6,7,8)
print(t[:])
print(t[3:])
print(t[:4])
print(t[-2:-5])   # step value by default positive
```

```python
t = (10,20,30,40,50,60,70,80)
print(t[1::2])
print(t[-1:-5:-2])
print(t[:: -1])
print(t[:7:2])
```
