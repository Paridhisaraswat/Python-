```python
# reverse a num
n=int(input("enter a num :"))
i = n
count = 0
rev = 0
p = 1

while( i != 0):
  i //=10
  count += 1

while(n != 0):
  a = n % 10
  p = 10 **(count - 1)
  ld= a * p
  rev += ld

  count -= 1
  n //= 10

print("reversed num :",rev)
```

```python
#armstrong num
sum = 0
n=int(input("enter a num :"))
i = n
while(i != 0):
  n1 = i % 10
  p = n1 ** 3
  sum += p
  i //= 10
if(sum == n):
  print("it is a armstrong num :",n)
else:
  print("not a Armstrong num ")
```

```python
#strong
sum = 0

n = int(input("enter a num :"))
i = n
while(i != 0):
  r = i % 10
  fact = 1
  while(r != 0):
    fact *= r
    r -= 1

  sum += fact
  i//= 10

if(sum == n):
  print("it is a strong num :",n)
else:
  print("not a strong num ")



```

```python
# print the list of anagrams
l = ["eat", "tea", "tan", "ate", "nat", "bat"]
anagrams = []
for i in range(len(l)):
  for j in range(i+1,len(l)):
    str1 = l[i]
    str2 = l[j]
    if(len(str1) == len(str2) ):
      count = 0
      for n in range(len(str1)):
          for m in range(len(str2)):

            if(str1[n] == str2[m]):
              count+=1
      if(count == len(str1)):

        if([str1,str2] not in anagrams):

          anagrams.append([str1,str2])

print(anagrams)
```
