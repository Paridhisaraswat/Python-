```python
#if a given year is a leap year. A leap year is divisible by 4 but not by 100 unless it is also divisible by 400.
year = int(input("enter a yaer :"))

if(year % 4 == 0 and year % 100 != 0 and year % 400 ==0 ):
    print("the entered year is a leap year",year)
else:
    print("the entered year is not a leap year")
```

```python
#WAP that calculates the Body Mass Index (BMI)
h=float(input("enter height:"))
height = h * 0.3048
weight=float(input("enter weight(in kg) :"))
bmi= weight / (height ** 2)

if(bmi < 18.5):
  print("underweight")
elif(18.5 <= bmi < 24.9):
  print("normal weight")
elif(25 <= bmi < 29.9):
  print("over weight")
else:
  print("obesity")
```

```python
'''WAP that calculates the discount for a product based on its price:
If price > 1000 , discount is 10%
If price is between 500 and 1000, discount is 5%
Otherwise, no discount
Print the final price after applying the discount.
'''

price=float(input("enter price :"))
if(price > 1000):
  print("the final price : ",price - (price * 0.1))
elif( 1000 >= price > 500):
  print("the final price :", price - (price * .05))
else:
  print("no discount ,",price )
```

```python
#WAP that simulates a simple ATM.
balance = 10000
choice= int(input("enter a choice(1-3) :"))

if(choice == 1):
  print("your current balance is :",balance )

elif(choice == 2):
  deposit=int(input("enter the amt to be deposited :"))
  if(deposit > 0):
    print("deposit successfully")
    balance+= deposit
    print("your balance after deposit is :", balance )
elif(choice == 3) :
  withdraw=int(input("enter the amt to be withdrawn"))
  if(withdraw > balance):
    print("insufficient fund")
  else:
    balance-=withdraw
    print("your balance after withdraw is :", balance)
else:
  print("enter a valid choice")
```
